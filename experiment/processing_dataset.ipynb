```python
import json
```

```python
with open("./dataset.json", "r") as jsonFile:
    ds = json.load(jsonFile) 
```

```python
len(ds)
```

```python
processed_ds = []

for i in ds:
    if isinstance(i, dict) and list(i.keys())[0] == "phrase":
        processed_ds.append({
            "text": i['phrase'], 
            "label": i['label']
        })

    elif isinstance(i, dict) and list(i.keys())[0] == "text":
        print(i)

        if len(list(i.keys())) == 1:
            print(list(i.keys()))
            if "Misselling" in i['text']:
                processed_ds.append({
                    "text": i['text'].replace("Misselling", ""), 
                    "label": "Misselling"
                })
            if "Not misselling" in i['text']:
                processed_ds.append({
                    "text": i['text'].replace("Not misselling", ""), 
                    "label": "Not misselling"
                })

            
        else:
            processed_ds.append({
                "text": i['text'], 
                "label": i['label']
            })

    

    elif isinstance(i, list) and len(i) == 2:
        processed_ds.append({
            "text": i[0], 
            "label": i[1]
        })
    
    elif isinstance(i, list) and len(i) == 1:
        if "misselling" in i[0].lower():
            processed_ds.append({
                "text": i[0].replace("Misselling", ""), 
                "label": "Misselling"
            })
        if "Not misselling" in i[0]:
            processed_ds.append({
                "text": i[0].replace("Not misselling", ""), 
                "label": "Not misselling"
            })
```

```python
with open("processed_dataset.json", "w") as jsonFile:
    json.dump(processed_ds, jsonFile, indent=4)
```

```python
import pandas as pd

df = pd.DataFrame(processed_ds)
```

```python
df = df.drop_duplicates()
```

```python
df['label'].value_counts()
```

```python
df['label'] = df['label'].replace({'not misselling': "Not Misselling", "misselling": "Misselling", 'Not misselling': "Not Misselling"})
```
