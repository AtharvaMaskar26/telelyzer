```python
from silero_vad import load_silero_vad, read_audio, get_speech_timestamps
model = load_silero_vad()
wav = read_audio('D:\\telelyzerv2\\experiment\\exp_data\\53501fef-ef8b-4a08-b171-bcd7846ea639.mp3')
speech_timestamps = get_speech_timestamps(
  wav,
  model,
  return_seconds=True,  # Return speech timestamps in seconds (default is samples)
)
```

```python
def convert_seconds_to_minutes(timestamp: float) -> str:
    seconds = int(timestamp % 60)
    minutes = int(timestamp / 60)
    hours = int(minutes / 60)

    return f"{hours}:{minutes}:{seconds}"

def display_timestamps_in_minutes(timestamp: dict) -> dict:
    start_time = convert_seconds_to_minutes(timestamp['start'])
    end_time = convert_seconds_to_minutes(timestamp['end'])

    return {
        'start': start_time, 
        'end': end_time
    }
```

```python
# Long pauses

for i in range(1, len(speech_timestamps)):
    if speech_timestamps[i]['start'] - speech_timestamps[i-1]['end'] >= 4:
        print(f"{display_timestamps_in_minutes(speech_timestamps[i-1])}-{display_timestamps_in_minutes(speech_timestamps[i])}")
```
