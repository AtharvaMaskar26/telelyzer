## Generate misselling dataset

````python
from openai import OpenAI
client = OpenAI()

def generate_synthetic_data():
    system_prompt = """
    Hey I am working on a product called telelyzer. It is a sales analysis tool that analyzes customer care call recordings. 
    Features of telelyzer are:
    1. Multi-lingual support for indian languages including Hindi, South Indian languages, Bengali BUT NOT Orissa, assamese, kashmiri. 
    2. Detect language switching, text sentiment analysis, scrip adherence and misselling. 
    3. Detect Voice Quality, audio dropouts, etc. 
    4. We are also working on Emotion analysis 
    5. Rate agent performance based on conversion rate, empathy etc. 

    I have given this product to sales agents to sell it, I wanna make sure they do not oversell or missell the product. 
    Help me generate synthetic dataset for finetuning BERT on binary classification to detect misselling. Try to cover various scenarios based on the features given. Return the result 10k in an array of labelled phrases with miselling or not misselling. 
    """


    completion = client.chat.completions.create(
    #   model="gpt-4o",
    model = "gpt-3.5-turbo",
    temperature = 0.3,
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": f"Genereate synthetic data and return an 100 synthetic entires to detect misselling. Just return the array and nothing else no extra text. Make sure every entry is unique and covers various variations that will make the model robust. Do not return ```python or ```"}
    ]
    )

    result = completion.choices[0].message.content

    return result
````

````python
final_dataset = []
````

````python
import ast
# final_dataset = []

for i in range(0, 300):
    try:
        i = generate_synthetic_data()
        final_dataset.append(ast.literal_eval(i))
    except Exception as e:
        pass
````
